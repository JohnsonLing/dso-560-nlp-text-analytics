#### To confirm that most words are taken out

```python
# https://medium.com/@cristhianboujon/how-to-list-the-most-common-words-from-text-corpus-using-scikit-learn-dad4d0cab41d
vec = CountVectorizer(stop_words="english", binary=True).fit(reviews_sample["Reviews"])
bag_of_words = vec.transform(reviews_sample["Reviews"])
sum_words = bag_of_words.sum(axis=0)
words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
words_freq[0:50]
```

#### Model 1: TFIDF vectorizer, no stopword removal or other preprocessing

```python
vectorizer = TfidfVectorizer(token_pattern=r'\b[a-zA-Z]{3,}\b')

corpus = list(reviews_sample["Reviews"].values)

X = vectorizer.fit_transform(corpus)
terms = vectorizer.get_feature_names()
tf_idf = pd.DataFrame(X.toarray(), columns = terms)
print(f"Shape of dataframe is {tf_idf.shape}")
tf_idf.head()
```

#### Model 2: Count vectorizer, no stopword removal or other preprocessing

```python
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(reviews_sample["Reviews"])

vectorized_df = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names())
print(f"Shape of dataframe is {vectorized_df.shape}")
print(f"Total number of occurences: {vectorized_df.sum().sum()}")
# print(f"Word counts: {vectorized_df.sum()}")
vectorized_df.head()
```

#### Model 3: Either Vectorizer + Stop Words Removed - TFIDF

```python
vectorizer = TfidfVectorizer(token_pattern=r'\b[a-zA-Z]{3,}\b', stop_words=stopwords.words())

corpus = list(reviews_sample["Reviews"].values)

X = vectorizer.fit_transform(corpus)
terms = vectorizer.get_feature_names()
tf_idf = pd.DataFrame(X.toarray(), columns = terms)
print(f"Shape of dataframe is {tf_idf.shape}")
tf_idf.head()
```

#### Model 4: Either Vectorization + Stemming 

```python
# https://www.projectpro.io/recipes/use-porter-stemmer
# https://www.simplilearn.com/tutorials/python-tutorial/list-to-string-in-python

stemmer = PorterStemmer()

list_with_stemmed = []
for i in range(len(reviews_sample)):
    tokenized_list = []
    stemmed_list = []
    tokenized_list.append(nltk.word_tokenize(reviews_sample["Reviews"].values[i]))
    for i in range(len(tokenized_list[0])):
        stemmed_list.append(stemmer.stem(tokenized_list[0][i]))
    list_with_stemmed.append(' '.join(stemmed_list))

data_with_stemmed = pd.DataFrame(list_with_stemmed, columns = ["Reviews"])
```

```python
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(data_with_stemmed["Reviews"])

vectorized_df = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names())
print(f"Shape of dataframe is {vectorized_df.shape}")
print(f"Total number of occurences: {vectorized_df.sum().sum()}")
#print(f"Word counts: {vectorized_df.sum()}")
vectorized_df.head()
```

#### Model 5: Either Vectorization + Lemmitazation

```python
rows = []
new_list = []
for i in range(len(reviews_sample)):
    doc = nlp(list(reviews_sample["Reviews"].values)[i])
    lemmatized_list = []
    for token in doc:
        rows.append((token.text, token.lemma_, token.pos_, token.tag_, token.dep_, token.shape_, token.is_alpha, token.is_stop))
        lemmatized_list.append(token.lemma_)
    lemmatized_list = ' '.join(lemmatized_list)
    new_list.append(lemmatized_list)

lemma_data = pd.DataFrame(new_list, columns = ["Reviews"])

data = pd.DataFrame(rows, columns=["text", "lemma", "part_of_speech", "tag", "dependency", "shape", "is_alphanumeric", "is_stopword"])
data.head()
```

```python
vectorizer = TfidfVectorizer(token_pattern=r'\b[a-zA-Z]{3,}\b')

corpus = list(lemma_data["Reviews"].values)

X = vectorizer.fit_transform(corpus)
terms = vectorizer.get_feature_names()
tf_idf = pd.DataFrame(X.toarray(), columns = terms)
print(f"Shape of dataframe is {tf_idf.shape}")
tf_idf.head()
```

#### Model 6: Count Vectorization with ngrams = 2

```python
vectorizer = CountVectorizer(ngram_range = (2, 2), stop_words="english", binary=True)
X = vectorizer.fit_transform(reviews_sample["Reviews"])

vectorized_df = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names())
print(f"Shape of dataframe is {vectorized_df.shape}")
print(f"Total number of occurences: {vectorized_df.sum().sum()}")
#print(f"Word counts: {vectorized_df.sum()}")
vectorized_df.head()
```

#### Model 7: TFIDF Vectorization with ngrams = 3

```python
vectorizer = TfidfVectorizer(ngram_range=(3,3),
                             token_pattern=r'\b[a-zA-Z]{3,}\b')

corpus = list(reviews_sample["Reviews"].values)

X = vectorizer.fit_transform(corpus)
terms = vectorizer.get_feature_names()
tf_idf = pd.DataFrame(X.toarray(), columns = terms)
print(f"Shape of dataframe is {tf_idf.shape}")
tf_idf.head()
```
